# 01 - Data Ingestion
In this notebook, we will load every dataset from the `data/raw` directory and perform initial exploratory data analysis. We will display the shape, columns, data types, summary statistics, and check for missing values.

```python
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

RAW_DIR = Path('../data/raw')
print(f'Raw Data Directory: {RAW_DIR}')
```

## 1. Locate Raw Datasets
We will find all `.csv` files inside the `data/raw` directory.

```python
raw_files = list(RAW_DIR.glob('*.csv'))
print(f'Found {len(raw_files)} files:')
for f in raw_files:
    print(f' - {f.name}')
```

## 2. Ingestion and Inspection Function
Let's define a function that loads a dataset and prints out its shape, columns, data types, missing values, and summary statistics.

```python
def inspect_dataset(file_path):
    print('='*50)
    print(f'Inspecting Dataset: {file_path.name}')
    print('='*50)
    
    df = pd.read_csv(file_path)
    print(f'\n[1] Shape: {df.shape}')
    print(f'\n[2] Columns: {list(df.columns)}')
    print(f'\n[3] Data Types:\n{df.dtypes}')
    print(f'\n[4] Missing Values:\n{df.isnull().sum()}')
    print(f'\n[5] Summary Statistics:\n{df.describe(include="all")}')
    print('\n')
    return df
```

## 3. Execute Inspection
We will now loop through all raw datasets and apply our inspection function.

```python
datasets = {}
for file_path in raw_files:
    try:
        datasets[file_path.name] = inspect_dataset(file_path)
    except Exception as e:
        print(f'Error reading {file_path.name}: {e}')
```
